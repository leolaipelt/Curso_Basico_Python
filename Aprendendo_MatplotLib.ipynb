# Aprendendo MatplotLib

A biblioteca MatplotLib é ideal para criar gráficos, plotar séries temporais.

O primeiro passo é importar a biblioteca no código para o Python reconhecer seus comandos:

```python
import matplotlib as plt
```

import -> importar a bliboteca
as -> um "apelido para facilitar o manuseio da biblioteca no código"

Por convenção utilizamos "plt" para o Matplotlib

Para visualizarmos os gráficos imediatamente após os códigos, precisamos avisar o notebook através da linha:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4])
plt.ylabel('Meu primeiro eixo Y')
plt.xlabel('Meu primeiro eixo X')
plt.title('Meu primeiro grafico com Python')
```

Utilizando a biblioteca numpy, podemos utilizar algumas ferramentas matemáticas no nosso gráfico

```python
import numpy as np
```

```python
exemplo=np.arange(0.,5.,0.1)
```

```python
exemplo #visualizando os números da array exemplo 
```

```python
plt.plot(exemplo)
```

Podemos criar multiplos gráficos em uma figura através da função subplot! 


```python
plt.figure(1) #criando o ambiente de plotagem

plt.subplot(211)
plt.plot(exemplo)

A=[5,4,3,2,1] #exemplo de array

plt.subplot(212)
plt.plot(A,'r')  #O comando 'r' dentro do plot significa a cor vermelha!
```

Um exemplo de criação de um hidrograma:

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
