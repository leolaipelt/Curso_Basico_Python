```
import pandas as pd
import folium
```

```
df=pd.read_csv('Rede_Hidrometeorológica_Nacional.csv')
```

```
df.head()
```

```
lat=df['Y'][:1000]
lon=df['X'][:1000]

MeuMapa=folium.Map(location=[-30,-51],zoom_start=2)

for lat_map, lon_map in zip(lat,lon):
  folium.Marker([lat_map,lon_map]).add_to(MeuMapa)
  
MeuMapa
```

```
from folium import plugins
coordenadas=[]
lat=df['Y'][:1000]
lon=df['X'][:1000]

MeuMapa = folium.Map(location=[-30,-51],zoom_start=4)

for la,lo in zip(lat,lon):
    coordenadas.append([la,lo])

MeuMapa.add_child(plugins.HeatMap(coordenadas))
MeuMapa
```
